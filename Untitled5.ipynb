Bütün sorulara tek yanıt veren öğrencilerin toplamı tüm öğrencilerin %51 i ne denk gelmektedir ve tüm sorulara aynı yanıtı veren öğrenciler en çok  3(nötr) yanıtını vermiştir . Bu göz ardı edilemeyecek kadar büyük olan ve artık "tekdüze" olarak nitelendireceğimiz yeni bir grubu ortaya çıkartır ve bu değerler modeli çok ciddi şekilde etkiyeceğinden her soruya aynı yanıtı veren öğrencilerle değişken yanıtlar veren   öğrenciler ayrı ayrı ve birbirleriyle değerlendirilecek aynı zamanda bu duruma ne neden olmuş olabilir bunun da cevabı aranacaktır. Bunun yanında Tekdüze yanıt veren öğrencilerin alt gruplarında tüm sorulara 1 verenlerin dersi daha kolay algıladığı ve diğer alt grupların zorluk algılarının birbirlerinden kayda değer derecede farklılık göstermemekte olduğu gözlenmiştir.  

```python

# Her bir ders için 'difficulty' dağılımını gösteren histogramlar
plt.figure(figsize=(15, 10))
for i, class_val in enumerate(df['class'].unique()):
    plt.subplot(3, 5, i + 1)
    sns.histplot(df[df['class'] == class_val]['difficulty'], bins=5, kde=True)
    plt.title(f'Ders {class_val}')
    plt.xlabel('Zorluk Seviyesi')
    plt.ylabel('Öğrenci Sayısı')
plt.tight_layout()
plt.show()

# Her bir ders için 'difficulty' dağılımını gösteren kutu grafikleri
plt.figure(figsize=(15, 8))
sns.boxplot(x='class', y='difficulty', data=df)
plt.title('Derslere Göre Zorluk Seviyesi Dağılımı')
plt.xlabel('Ders Kodu')
plt.ylabel('Zorluk Seviyesi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Tekdüze verilere herhangi bir uygulama yapmadan önce her ders ile ayrı ayrı  zorluk seviyelerinin dağılımı gösteren grafikler yukarıdadır. Bu grafiklere göre derslerin büyük çoğunluğunda zorluk 3 yani orta düzeyde seyrederken bazı derslerde (ders 13 , 3 , 4 , 5 ve 9 ) zorluğun 1 olduğunu düşünlerin sayısı daha fazladır . Bunun yanında ders 7 nin zorluk derecesinin diğer derslere göre daha fazla olduğu dikkat çekmektedir.  Ders 12 de çok az sayıda veri olduğu ve güvenirliğinin sorgulanması gerketiği unutulmamalıdır

```python
import numpy as np
question_cols = [f'Q{i}' for i in range(1, 29)]
df['response_std'] = df[question_cols].std(axis=1)
is_sl = df['response_std'] == 0
df_sl = df[is_sl].copy()
df_dy = df[~is_sl].copy() # is_sl olmayanlar

print(f"Toplam Öğrenci/Yanıt Sayısı: {len(df)}")
print(f"Tekdüze Yanıt Verenlerin  Sayısı: {len(df_sl)}")
print(f"Farklı Değerler Verenlerin Sayısı: {len(df_dy)}")
def plot_difficulty_per_class(dataframe, group_title):
    """
    Verilen DataFrame ve grup başlığı için derslere göre zorluk dağılımını
    histogram ve kutu grafikleriyle çizer.
    """
    if dataframe.empty:
        print(f"'{group_title}' grubu için veri bulunamadı, grafikler çizilmiyor.")
        return

    print(f"\n--- {group_title} Grubu İçin Grafikler Oluşturuluyor ---")


    # Bu gruptaki mevcut ders kodlarını alalım ve sıralayalım
    unique_classes = sorted(dataframe['class'].unique())
    n_classes = len(unique_classes)

    if n_classes == 0:
        print(f"'{group_title}' grubu için ders bulunamadı.")
        return

    #  Histogramlar

    ncols = 5
    nrows = int(np.ceil(n_classes / ncols))

    fig_hist, axes_hist = plt.subplots(nrows, ncols, figsize=(15, 3 * nrows), squeeze=False) # squeeze=False önemli
    fig_hist.suptitle(f'{group_title}: Derslere Göre Zorluk Dağılımı (Histogram)', fontsize=16, y=1.02)

    axes_flat = axes_hist.flatten() # Subplot eksenlerini düzleştir
    for i, class_val in enumerate(unique_classes):
        ax = axes_flat[i]
        class_data = dataframe[dataframe['class'] == class_val]['difficulty']
        if not class_data.empty:
            sns.histplot(class_data, bins=5, kde=True, ax=ax, stat='count') # Öğrenci sayısını göstermek için stat='count'
            ax.set_title(f'Ders {class_val}')
            ax.set_xlabel('Zorluk Seviyesi')
            ax.set_ylabel('Öğrenci Sayısı')
        else:
            ax.set_title(f'Ders {class_val} (Veri Yok)')
            ax.set_axis_off() # Veri yoksa eksenleri kapat

    # Kullanılmayan fazla subplotları gizle
    for j in range(i + 1, nrows * ncols):
        axes_flat[j].set_axis_off()

    plt.tight_layout(rect=[0, 0, 1, 0.98]) # Ana başlık için yer bırak
    plt.show()

    # Kutu Grafikleri
    plt.figure(figsize=(15, 8))
    # order parametresi ile derslerin sıralamasını sabit tutalım
    sns.boxplot(x='class', y='difficulty', data=dataframe, order=unique_classes)
    plt.title(f'{group_title}: Derslere Göre Zorluk Seviyesi Dağılımı (Kutu Grafik)')
    plt.xlabel('Ders Kodu')
    plt.ylabel('Zorluk Seviyesi')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

#  Her Grup İçin Grafikleri Çizdirme
plot_difficulty_per_class(df_sl, "Straight-Liners (Tüm Yanıtları Aynı Olanlar)")
plot_difficulty_per_class(df_dy, "Variable Responders (Farklı Yanıtlar Verenler)")
```

Veri setini tüm cevaplara aynı yanıtı veren "tekdüze" değerler ile farklı yanıtlar veren öğrencilerin derslerin ayrı ayrı zorluk dağımı grafiklerine bakarak sorulara değişken yanıtlar veren öğrencilerin derslerin genelinde zorluk seviyesini 3 olarak düşündükleri bunun yanında  tekdüze yanıt veren öğrencilerde zorluk seviyeleri derslerin genelinde uç zorluk değerlerine (bazı derslerde en kolay olan 1  bazı derslerde ise en zor olan 5'e ) daha yakın  olduğunu gözlemliyoruz

```python

# 'repeat' ve 'difficulty' arasındaki ilişkiyi gösteren kutu grafiği
plt.figure(figsize=(8, 6))
sns.boxplot(x='nb.repeat', y='difficulty', data=df)
plt.title('Tekrar Sayısına Göre Zorluk Seviyesi Dağılımı')
plt.xlabel('Dersi Kaçıncı Kez Aldığı')
plt.ylabel('Zorluk Seviyesi')
plt.show()
```

```python

print("\n--- SL ve DY Grupları İçin Genel Zorluk Dağılımı Grafikleri ---")

# --- Veri Hazırlığı ve Kontroller ---
sl_counts_plot = pd.Series(dtype=int)
dy_counts_plot = pd.Series(dtype=int)

if 'df_sl' not in locals() or df_sl.empty:
    print("SL grubu (df_sl) boş veya tanımlı değil.")
else:
    # SL grubu için zorluk seviyesi sayıları
    sl_counts = df_sl['difficulty'].value_counts()
    # Grafikte 1-5 arası tüm barları görmek için reindex kullanalır
    sl_counts_plot = sl_counts.reindex(range(1, 6), fill_value=0).sort_index()

if 'df_dy' not in locals() or df_dy.empty:
    print("DY grubu (df_dy) boş veya tanımlı değil.")
else:
    # DY grubu için zorluk seviyesi sayıları
    dy_counts = df_dy['difficulty'].value_counts()
    # Grafikte 1-5 arası tüm barları görmek için reindex kullanalım
    dy_counts_plot = dy_counts.reindex(range(1, 6), fill_value=0).sort_index()

# --- Karşılaştırmalı Bar Grafiği ---
# Eğer her iki grup için de veri varsa çizelim
if not sl_counts_plot.empty or not dy_counts_plot.empty:
    fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True) # Y eksenini paylaşmak karşılaştırmayı kolaylaştırır
    fig.suptitle('Genel Zorluk Seviyesi Dağılımı (SL vs DY)', fontsize=16, y=1.02)

    # SL Grubu Grafiği
    if not sl_counts_plot.empty:
        axes[0].bar(sl_counts_plot.index, sl_counts_plot.values, color='lightcoral')
        axes[0].set_title('SADECE SL Grubu')
        axes[0].set_xlabel('Zorluk Seviyesi (1-5)')
        axes[0].set_ylabel('Öğrenci Sayısı')
        axes[0].set_xticks(range(1, 6)) # X ekseninde 1-5 arası etiketler
        axes[0].grid(axis='y', linestyle='--', alpha=0.7)
        # Barların üzerine sayıları yazdır
        for i, v in enumerate(sl_counts_plot):
            axes[0].text(i + 1, v + sl_counts_plot.max()*0.01, str(v), ha='center', va='bottom') # X ekseni index+1 olmalı
    else:
        axes[0].set_title('SADECE SL Grubu (Veri Yok)')
        axes[0].set_axis_off()

    # DY Grubu Grafiği
    if not dy_counts_plot.empty:
        axes[1].bar(dy_counts_plot.index, dy_counts_plot.values, color='skyblue')
        axes[1].set_title('SADECE DY Grubu')
        axes[1].set_xlabel('Zorluk Seviyesi (1-5)')
        # axes[1].set_ylabel('Öğrenci Sayısı') # Y ekseni paylaşıldığı için tekrar yazmaya gerek yok
        axes[1].set_xticks(range(1, 6))
        axes[1].grid(axis='y', linestyle='--', alpha=0.7)
        # Barların üzerine sayıları yazdır
        for i, v in enumerate(dy_counts_plot):
            axes[1].text(i + 1, v + dy_counts_plot.max()*0.01, str(v), ha='center', va='bottom')
    else:
        axes[1].set_title('SADECE DY Grubu (Veri Yok)')
        axes[1].set_axis_off()

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ana başlık için yer ayarla
    plt.show()

    # Sayısal Değerleri de Görelim
    print("\nSL Grubu Zorluk Dağılımı (Sayısal):")
    print(sl_counts_plot)
    print("\nDY Grubu Zorluk Dağılımı (Sayısal):")
    print(dy_counts_plot)

else:
    print("Grafik çizdirmek için SL veya DY gruplarından en az birinde veri olması gerekli.")
```

Bu karşılaştırmalı grafik, SL ve DY gruplarının derslerin zorluğunu algılama biçimlerinde sistematik farklılıklar olduğunu güçlü bir şekilde gösteriyor.DY grubu, (Değişken yanıtlar) zorluğu daha çok beklendiği gibi orta (3) etrafında algılama eğiliminde.

SL grubu ise (Tekdüze Yanıtlar) daha polarize bir algıya sahip görünüyor; özellikle çok kolay/kolay (1) algısı bu grupta son derecede yaygın, bunun yanında orta (3) algısı da oldukça güçlü. Bu durum, SL1 alt grubunun büyüklüğünü ve SL3 alt grubunun da etkisini yansıtıyor olabilir. SL grubundaki yüksek '1' oranı, belki de anketi hızlı geçiştirirken zorluk sorusuna da '1' yanıtını verme eğilimiyle de ilişkili olabilir.

```python

# is_straightliner sütununu oluştur (True/False)
df['is_straightliner'] = (df['response_std'] == 0)
print("SL ve DY Gruplarının Karşılaştırılması (Genel Bakış)")
print(df['is_straightliner'].value_counts()) # Sayıları kontrol et
# --- 2. Genel Devam Durumu Farkı (Düzeltilmiş Kod) ---
print("\n--- 2. Devam Durumu Karşılaştırması (SL vs DY) ---")

# 'attendance' sütununun varlığını kontrol et
if 'attendance' not in df.columns:
    print("HATA: 'attendance' sütunu DataFrame'de bulunamadı.")
else:
    # Çapraz Tablo (Oranlar)
    # dropna=False ile attendance NaN ise onu da ayrı bir kategori olarak görebiliriz
    attendance_crosstab_norm = pd.crosstab(df['is_straightliner'], df['attendance'], normalize='index', dropna=False)

    # İndeks etiketlerini daha anlaşılır yapalım
    attendance_crosstab_norm.index = attendance_crosstab_norm.index.map({True: 'SL Grubu', False: 'DY Grubu'})

    print("Devam Seviyesi Oranları (Her grubun kendi içindeki %):")
    pd.set_option('display.float_format', '{:.3f}'.format) # Ondalık formatı ayarla
    print(attendance_crosstab_norm)

    # Görselleştirme (Yığılmış Bar Grafik)
    # Eğer crosstab boşsa (örn: attendance sütunu tamamen NaN ise) hata vermemesi için kontrol
    if not attendance_crosstab_norm.empty:
        ax = attendance_crosstab_norm.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
        plt.title('Devam Seviyesi Dağılımı (SL vs DY)')
        plt.xlabel('Öğrenci Grubu')
        plt.ylabel('Öğrenci Oranı')
        plt.xticks(rotation=0) # X ekseni etiketlerini düz yap
        # Legend'i dışarı taşıma (isteğe bağlı)
        plt.legend(title='Devam Seviyesi (0-4)', bbox_to_anchor=(1.02, 1), loc='upper left')
        plt.tight_layout(rect=[0, 0, 0.85, 1]) # Legend için yer ayır
        plt.show()
    else:
        print("Devam durumu için çapraz tablo oluşturulamadı (muhtemelen 'attendance' sütununda sorun var).")
```

Yukardaki devam seviyesine göre öğrenci oranı karşılaştırmalı  grafiğine göre tekdüze yanıt veren öğrenciler derslere belirgin bir şekilde değişken yanıt verenlerden daha az katılım gösteriyor.

```python
# makine öğrenmesi tekniklerinin içeri aktarımı
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
import warnings

#  Özellik Mühendisliği
def group_difficulty(difficulty):
    if difficulty <= 2: # 1 ve 2
        return 0 # Memnun Değil
    elif difficulty == 3: # 3
        return 1 # Orta
    else: # 4 ve 5
        return 2 # Memnun
# zorluk derecelerini gruplandırarak hedef değişkeni zorlukların sınıfları olan bir sınıflandırma problemine dönüştürdük
df['difficulty_grouped'] = df['difficulty'].apply(group_difficulty)

# Hedef değişken (y) ve özellikler (X)
target_column = 'difficulty_grouped'
# Orijinal 'difficulty' sütununu özelliklerden çıkaralım
features = df.drop(['difficulty', target_column], axis=1)
X = features
y = df[target_column]

print("\n'difficulty' gruplandıktan sonra sınıf dağılımı:")
print(y.value_counts(normalize=True) * 100) # Yüzdesel dağılımı görmek için normalize=True
# gruplandırılmış zorluk değerlerininin sınıfsal dağılımı


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
# 'stratify=y' parametresi, eğitim ve test setlerindeki sınıf oranlarının orijinal veri setindekine yakın olmasını sağlar.
# Bu, özellikle dengesiz sınıflarda önemlidir.

print(f"\nEğitim seti boyutu: {X_train.shape}, {y_train.shape}")
print(f"Test seti boyutu: {X_test.shape}, {y_test.shape}")

# Ölçeklendirme
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Model Seçimi
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine": SVC(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42)
}

#  Model Eğitimi ve Değerlendirme
results = {}

print("\n--- Model Eğitimi ve Değerlendirme ---")
for name, model in models.items():
    print(f"\nModel: {name}")

    # Modeli eğitme
    model.fit(X_train_scaled, y_train)

    # Test seti üzerinde tahmin yapma
    y_pred = model.predict(X_test_scaled)

    # Doğruluğu hesaplama
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f"Doğruluk (Accuracy): {accuracy:.4f}")

    # Sınıflandırma Raporu
    print("Sınıflandırma Raporu:")

    target_names = ['Kolay (0)', 'Orta (1)', 'Zor (2)']
    print(classification_report(y_test, y_pred, target_names=target_names))

print("\n--- Sonuçların Karşılaştırılması ---")
# Sonuçları doğruluk oranına göre sıralama
sorted_results = dict(sorted(results.items(), key=lambda item: item[1], reverse=True))
for name, accuracy in sorted_results.items():
    print(f"{name}: {accuracy:.4f}")
```

Tüm öğrencilerle yapılan ilk değerlendirmelerde, memnuniyet verileri kullanılarak ders zorluk algısının tahmininde en yüksek doğruluk oranı  %58 olarak kaydedilmiştir. Tüm verinin yarısını oluşturan tekdüze yanıt veren öğrencilerin modele olan etkisini ölçmek için sadece değişken yanıt veren öğrencilerle tekrar deneme yapıyoruz.

```python
import pandas as pd
import numpy as np

# GENEL BİR MEMNUNİYET SKORU HESAPLAYIP BU SKORU BİR ÖNCEKİ MODELDEKİ GİBİ KATEGORİLERE AYIRIYORUZ


q_columns = [f'Q{i}' for i in range(1, 29)]

# Overall_Score'u hesapla (sadece mevcut Q sütunlarının ortalaması)

if q_columns: # Eğer Q sütunu kaldıysa hesapla
    df_dy['Overall_Score'] = df_dy[q_columns].mean(axis=1)
    print("'Overall_Score' hesaplandı ve df_dy'ye eklendi.")
    print(df_dy[['Overall_Score']].head()) # İlk birkaç skoru göster
else:
    print("HATA: Ortalama hesaplamak için Q sütunları bulunamadı!")
    # Bu durumda kodu durdurmak veya başka bir skor tanımı yapmak gerekir.


#  Tercile Eşiklerini Belirleme
# bu overall_Score çıktılarını gruplandırırken dengeli bir dağılım oluşturmak içindir
if 'Overall_Score' in df_dy.columns:
    # 33.3 ve 66.7 yüzdelikleri (terciles) hesapla
    q33 = df_dy['Overall_Score'].quantile(0.3333)
    q67 = df_dy['Overall_Score'].quantile(0.6667)

    print(f"\nÜçtebirlik Dilim Eşikleri:")
    print(f" - Düşük Memnuniyet Sınırı (Alt Eşik): < {q33:.3f}")
    print(f" - Orta Memnuniyet Sınırı (Üst Eşik): < {q67:.3f}")


    #  Kategorize Etme Fonksiyonu
    def categorize_satisfaction_tercile(score, lower_bound, upper_bound):
        # Not: Eşitlik durumlarını dikkatli ele almak önemli

        if score < lower_bound:
            return 0 # Düşük Memnuniyet
        elif score < upper_bound:
            # Eğer skor tam olarak lower_bound'a eşitse, onu Orta gruba dahil ediyoruz.
            return 1 # Orta Memnuniyet
        else:
            # upper_bound'a eşit veya büyükse Yüksek gruba dahil ediyoruz.
            return 2 # Yüksek Memnuniyet

    # Kategorize etme fonksiyonunu uygula
    df_dy['satisfaction_category'] = df_dy['Overall_Score'].apply(
        lambda score: categorize_satisfaction_tercile(score, q33, q67)
    )
    print("\n'satisfaction_category' oluşturuldu ve df_dy'ye eklendi.")


    # Kategori Dağılımını Kontrol Etme
    print("\nMemnuniyet Kategorisi Dağılımı (Tercile Bazlı):")
    print(df_dy['satisfaction_category'].value_counts(normalize=True).sort_index() * 100)
    # Bu çıktıda yüzdelerin birbirine yakın (~%33 civarında) olması beklenir.


    #  Modelleme İçin Hedef Değişkeni Tanımlama
    y_satisfaction = df_dy['satisfaction_category']
    print("\nModelleme için 'y_satisfaction' hedef değişkeni tanımlandı.")
    print(y_satisfaction.head())

else:
    print("\n'Overall_Score' hesaplanamadığı için kategorizasyon yapılamadı.")
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
import warnings


try:
    y_train_satisfaction = y_satisfaction.loc[X_train_dy_selected.index]
    y_test_satisfaction = y_satisfaction.loc[X_test_dy_selected.index]
    print("y_satisfaction (hedef değişken), eğitim ve test setlerine başarıyla ayrıldı.")
    print(f"y_train_satisfaction boyutu: {y_train_satisfaction.shape}")
    print(f"y_test_satisfaction boyutu: {y_test_satisfaction.shape}")
    data_ready = True
except NameError as e:
    print(f"\n--- HATA ---")
    print(f"Gerekli değişken bulunamadı: {e}.")
    print("Lütfen bir önceki adımdaki kodların başarıyla çalıştığından ve")
    print("'X_train_dy_selected', 'X_test_dy_selected' ve 'y_satisfaction' değişkenlerinin")
    print("tanımlı olduğundan emin olun.")
    print("----------------------------------------------------------")
    data_ready = False
except Exception as e:
    print(f"\n--- HATA ---")
    print(f"y_satisfaction ayrılırken beklenmedik bir sorun oluştu: {e}")
    print("----------------------------------------------------------")
    data_ready = False


#  Model Seçimi
if data_ready:
    models_satisfaction = {
        "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
        "K-Nearest Neighbors": KNeighborsClassifier(),
        "Support Vector Machine": SVC(random_state=42),
        "Decision Tree": DecisionTreeClassifier(random_state=42),
        "Random Forest": RandomForestClassifier(random_state=42),
        "Gradient Boosting": GradientBoostingClassifier(random_state=42)
    }



    # Model Eğitimi ve Değerlendirme
    results_satisfaction = {}

    print("\n--- Model Eğitimi ve Değerlendirme (Memnuniyet Kategorisi Tahmini) ---")

    for name, model in models_satisfaction.items():
        print(f"\nModel: {name}")

        try:
            # Modeli eğit (SEÇİLMİŞ özellikler, YENİ memnuniyet hedefi ile)
            model.fit(X_train_dy_selected, y_train_satisfaction)

            # Test seti üzerinde tahmin yap (SEÇİLMİŞ özelliklerle)
            y_pred_satisfaction = model.predict(X_test_dy_selected)

            # Doğruluğu hesapla
            accuracy = accuracy_score(y_test_satisfaction, y_pred_satisfaction)
            results_satisfaction[name] = accuracy
            print(f"Doğruluk (Accuracy): {accuracy:.4f}")

            # Sınıflandırma Raporu
            print("Sınıflandırma Raporu:")
            # Kategorilere verdiğimiz isimleri tanımlayalım (0: Düşük, 1: Orta, 2: Yüksek varsayımıyla)
            satisfaction_target_names = ['Düşük Memnuniyet (0)', 'Orta Memnuniyet (1)', 'Yüksek Memnuniyet (2)']
            print(classification_report(y_test_satisfaction, y_pred_satisfaction, target_names=satisfaction_target_names, zero_division=0))

        except Exception as e:
            print(f"'{name}' modeli çalıştırılırken bir hata oluştu: {e}")
            results_satisfaction[name] = 0.0 # Hata durumunda skoru 0 yap

    print("\n--- Memnuniyet Tahmini Sonuçlarının Karşılaştırılması ---")
    # Sonuçları doğruluk oranına göre sırala
    sorted_results_satisfaction = dict(sorted(results_satisfaction.items(), key=lambda item: item[1], reverse=True))
    for name, accuracy in sorted_results_satisfaction.items():
        print(f"{name}: {accuracy:.4f}")

else:
    print("\nVeri hazırlama adımında hata oluştuğu için model eğitimi ve değerlendirme adımı atlandı.")
```

Bu sonuçlara göre Logistic Regression modeli %93.65 gibi çok yüksek bir test doğruluğu ile en iyi performansı sergilemiş, diğer algoritmalar da %90'ın üzerinde doğruluklarla yakından takip etmiştir.Decision Tree modeli ise %82.01 ile diğerlerine kıyasla daha düşük bir doğrulukta kalmıştır.Bu yüksek doğruluk oranlarının temel nedeni, tahmin etmeye çalıştığımız hedef değişkeninin tahmin için kullanılan özelliklerin ortalamasından türetilmiş olmasıdır.
```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay



try:
    if 'sorted_results_satisfaction' not in locals() or \
       not isinstance(sorted_results_satisfaction, dict) or \
       not sorted_results_satisfaction:
        raise NameError("'sorted_results_satisfaction' sözlüğü bulunamadı veya boş.")
    if 'models_satisfaction' not in locals() or \
       not isinstance(models_satisfaction, dict) or \
       not models_satisfaction:
        raise NameError("'models_satisfaction' sözlüğü bulunamadı veya boş.")
    if 'X_test_dy_selected' not in locals(): # DataFrame olup olmadığını kontrol etmek zor olabilir, sadece varlığını kontrol edelim
        raise NameError("'X_test_dy_selected' DataFrame'i bulunamadı.")
    if 'y_test_satisfaction' not in locals(): # Series olup olmadığını kontrol etmek zor olabilir
        raise NameError("'y_test_satisfaction' Serisi bulunamadı.")

    # En iyi modeli seç
    best_model_name_for_cm = list(sorted_results_satisfaction.keys())[0]
    best_model_for_cm = models_satisfaction[best_model_name_for_cm]

    print(f"Karışıklık matrisi için en iyi model seçildi: {best_model_name_for_cm}")
    print(f"Bu modelin test doğruluğu: {sorted_results_satisfaction[best_model_name_for_cm]:.4f}")

    #  model zaten eğitilmişti  Test seti üzerinde tekrar tahmin yapalım.
    y_pred_for_cm = best_model_for_cm.predict(X_test_dy_selected)

    # Karışıklık Matrisi
    #
    satisfaction_display_labels = ['Düşük Mem. (0)', 'Orta Mem. (1)', 'Yüksek Mem. (2)']

    cm = confusion_matrix(y_test_satisfaction, y_pred_for_cm)

    print(f"\n--- '{best_model_name_for_cm}' için Karışıklık Matrisi (Sayısal) ---")
    # Matrisi daha okunaklı yazdırmak için küçük bir formatlama
    header = "Gerçek ↓ | Tahmin →"
    class_headers = "".join([f"{label:<18}" for label in satisfaction_display_labels])
    print(f"{header:<18} {class_headers}")
    print("-" * (len(header) + len(class_headers) + 5))
    for i, row_label in enumerate(satisfaction_display_labels):
        row_values = "  ".join([f"{count:<18}" for count in cm[i]])
        print(f"{row_label:<18} {row_values}")

    # --- Karışıklık Matrisini Görselleştirme (Scikit-learn ile) ---
    print("\nKarışıklık Matrisi Görseli (Scikit-learn Display):")

    fig, ax = plt.subplots(figsize=(8, 6)) # Grafik boyutunu ayarla
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=satisfaction_display_labels)
    disp.plot(cmap=plt.cm.Blues, ax=ax, colorbar=True) # 'ax=ax' ile figür ve eksen belirtiyoruz

    plt.xticks(rotation=30, ha="right") # X ekseni etiketlerini eğik yap
    plt.title(f"Karışıklık Matrisi - {best_model_name_for_cm}", fontsize=14)
    plt.xlabel("Tahmin Edilen Memnuniyet Kategorisi", fontsize=12)
    plt.ylabel("Gerçek Memnuniyet Kategorisi", fontsize=12)
    plt.tight_layout() # Kenar boşluklarını ayarla
    plt.show()



except NameError as e:
    print(f"\nHATA: Gerekli bir değişken bulunamadı: {e}")
    print("Lütfen önceki memnuniyet modeli eğitimi ve değerlendirme adımlarının eksiksiz çalıştığından emin olun.")
    print("Özellikle 'models_satisfaction', 'X_test_dy_selected', 'y_test_satisfaction' ve 'sorted_results_satisfaction' değişkenleri gereklidir.")
except Exception as e:
    print(f"HATA: Karışıklık matrisi oluşturulurken veya gösterilirken bir sorun oluştu: {e}")
```
